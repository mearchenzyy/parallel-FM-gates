```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
import pandas as pd
```

```python
def read_data(file):
    """reads in Fourier_Pulse_Qm_Qn.txt file and outputs arrays of freq and amp"""
    data = pd.read_csv(file, header = None, sep="   ", engine='python') #seperator is a tab
    freq = (data[0])
    amp = data[2]
    return np.array(freq), np.array(amp)
    return data
```

```python
filename=["Fourier_Pulse_Q1_Q2.txt","Fourier_Pulse_Q4_Q5.txt"]
for file in filename:
    freq,amp = (read_data(file))
    plt.figure()
    plt.bar (freq,amp,width=0.004)
    plt.title(file)
    plt.ylabel('Amp')
    plt.xlabel('Freq (MHz)')
```
